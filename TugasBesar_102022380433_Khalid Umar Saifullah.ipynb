```python
import datetime
import math

# Database sederhana untuk menyimpan data parkir
data_parkir = {}
#password admin
secret_key = "1234"
# Tarif , denda, dan satuan waktu (bulat)
satuan_bulat = 60
tarif_per_detik = 10000
denda_4_menit = 0.10
denda_6_menit = 0.25
batas_pertama = 4
batas_kedua = 6

# Fungsi untuk pencatatan kendaraan masuk
def masuk_parkir():
    nomor_plat = input("Masukkan nomor plat kendaraan: ")
    waktu_masuk = datetime.datetime.now()
    data_parkir[nomor_plat] = {"waktu_masuk": waktu_masuk}
    print(data_parkir)
    print("Silahkan masuk.")

# Fungsi untuk pencatatan kendaraan keluar
def keluar_parkir():
    nomor_plat = input("Masukkan nomor plat kendaraan: ")
    if nomor_plat in data_parkir:
        waktu_keluar = datetime.datetime.now()
        biaya = hitung_biaya(nomor_plat, waktu_keluar)
        print(f"Biaya parkir: Rp {biaya}")
        while True:
            try:
                pembayaran = int(input("Masukkan nominal pembayaran: "))
                if pembayaran < biaya:
                    raise ValueError("Nominal pembayaran kurang dari biaya parkir. Silakan masukkan jumlah yang benar.")
                break
            except ValueError as e:
                print(e)
        print("Terima Kasih.")
        del data_parkir[nomor_plat]
    else:
        print("Kendaraan tidak terdaftar di area parkir.")

# Fungsi untuk menghitung biaya parkir
def hitung_biaya(nomor_plat, waktu_keluar):
    waktu_masuk = data_parkir[nomor_plat]["waktu_masuk"]
    durasi_detik = (waktu_keluar - waktu_masuk).total_seconds()
#     durasi_detik =380
    print(f"Durasi Detik :{durasi_detik}")
    if durasi_detik >= 60:
        if (durasi_detik%satuan_bulat !=0) and (durasi_detik%satuan_bulat >14):
            durasi_menit = math.ceil(durasi_detik/satuan_bulat)
        else:
            durasi_menit = math.floor(durasi_detik/satuan_bulat)
    elif durasi_detik <60:
        durasi_menit = math.ceil(durasi_detik/satuan_bulat)
    print(f"Durasi Menit:{durasi_menit}")
    biaya = durasi_menit * tarif_per_detik
    if durasi_menit > batas_pertama and durasi_menit <= batas_kedua:
        biaya += biaya * denda_4_menit
    elif durasi_menit > batas_kedua:
        biaya += biaya * denda_6_menit
    return float(biaya)

# Fungsi untuk menu admin
def menu_admin():
    pin = input("Masukkan PIN Admin: ")
    if pin == secret_key:  # Contoh PIN sederhana
        for nomor_plat, info in data_parkir.items():
            waktu_masuk_format = info['waktu_masuk'].strftime("%Y-%m-%d %H:%M:%S")
            print(f"Plat Nomor :{nomor_plat} | Waktu Masuk : {waktu_masuk_format}")
    else: 
        print("PIN salah!")

# Fungsi utama
def main():
    while True:
        try:
            print("\nMenu Utama Aplikasi Parkir")
            print("1. Masuk Area Parkir")
            print("2. Keluar Area Parkir")
            print("3. Admin Parkir")
            print("4. Keluar")
            pilihan = input("Pilih menu: ")

            if pilihan == "1":
                masuk_parkir()
            elif pilihan == "2":                                
                keluar_parkir()
            elif pilihan == "3":
                menu_admin()
            elif pilihan == "4":
                break
            else:
                raise ValueError("Pilihan tidak valid.")  # Membuat eksepsi untuk input tidak valid
        except ValueError as e:
            print(e)
# Jalankan program
if __name__ == "__main__":
    main()
```
